```python
# Lecture des fichiers
df_crypto_infos = pd.read_csv("csv/csv_bdd/crypto_infos.csv")
df_crypto_hist = pd.read_csv("csv/csv_bdd/historique_cryptos.csv")

# R√©cup√©rer les tickers uniques de chaque DataFrame
tickers_infos = set(df_crypto_infos["Ticker_cryptos_Yf"].unique())
tickers_hist = set(df_crypto_hist["Ticker_cryptos_Yf"].unique())

# Trouver les diff√©rences
tickers_manquants = tickers_infos - tickers_hist  # Dans infos mais pas dans hist
tickers_en_trop = tickers_hist - tickers_infos    # Dans hist mais pas dans infos

print(f"üìä Infos : {len(tickers_infos)} tickers")
print(f"üìä Historique : {len(tickers_hist)} tickers")

print(f"\n‚ùå Tickers dans infos MAIS PAS dans historique ({len(tickers_manquants)}):")
for ticker in sorted(tickers_manquants):
    print(f"  - {ticker}")

print(f"\n‚ö†Ô∏è Tickers dans historique MAIS PAS dans infos ({len(tickers_en_trop)}):")
for ticker in sorted(tickers_en_trop):
    print(f"  - {ticker}")
```
